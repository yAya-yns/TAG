```python
with open('../data/results_cifar10_extended.json', 'r') as f:
    results = json.load(f)
```

# Plot all training and evaluation curves

```python
plt.rcParams.update({'font.size': 16})

fig, axes = plt.subplots(3, 2, figsize=(12,10))

# val_acc and val_acc_noise denote results on the validation set (clean and noisy images) of CIFAR-10
# test_acc and test_acc_noise denote results on the test set (clean and noisy images) of CIFAR-10

for ind, attr in enumerate(['training_loss', 'training_acc', 
                            'val_acc', 'test_acc',
                            'val_acc_noise', 'test_acc_noise']):
    
    ax = axes[ind // 2][ind % 2]
    c = 0
    for split, values in results.items():
        for seed in values:
            if attr.startswith('training'):
                x = np.arange(len(values[seed][attr])) + 1
            else:
                x = [1, 5, 10, 20, 30, 40, 50]
            ax.plot(x, values[seed][attr], alpha=0.5)
            c += 1
    if attr == 'training_loss':
        ax.set_yscale('log')
    else:
        ax.set_ylim([5, 105])
    ax.set_xlabel('Epoch')
    ax.set_ylabel(attr)
    ax.grid(lw=0.5)
    ax.set_title('%s for %d models' % (attr, c))
plt.tight_layout(pad=2)
plt.show()
```
